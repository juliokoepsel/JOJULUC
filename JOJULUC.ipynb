#Análise Léxica:

```python
import ply.lex as lex

# Tokens
tokens = (
    'ID', 'INT', 'FLOAT', 'STRING', 'BOOL',
    'PLUS', 'MINUS', 'TIMES', 'DIVIDE', 'MOD',
    'GT', 'LT', 'EQ', 'NE', 'LE', 'GE',
    'AND', 'OR', 'NOT',
    'LPAREN', 'RPAREN', 'LBRACE', 'RBRACE',
    'COMMA', 'SEMI', 'ASSIGN',
    'FUNCTION', 'MAIN', 'IF', 'ELSE', 'LOOP', 'BREAK', 'RETURN', 'PRINT', 'TYPE'
)

# Palavras reservadas
reserved = {
    'function': 'FUNCTION',
    'main': 'MAIN',
    'if': 'IF',
    'else': 'ELSE',
    'loop': 'LOOP',
    'break': 'BREAK',
    'return': 'RETURN',
    'print': 'PRINT',
    'void': 'TYPE',
    'int': 'TYPE',
    'float': 'TYPE',
    'string': 'TYPE',
    'bool': 'TYPE',
    'true': 'BOOL',
    'false': 'BOOL'
}

# Expressão regulares para tokens simples
t_PLUS = r'\+'
t_MINUS = r'-'
t_TIMES = r'\*'
t_DIVIDE = r'/'
t_MOD = r'%'
t_GT = r'>'
t_LT = r'<'
t_EQ = r'=='
t_NE = r'!='
t_LE = r'<='
t_GE = r'>='
t_AND = r'&&'
t_OR = r'\|\|'
t_NOT = r'!'
t_LPAREN = r'\('
t_RPAREN = r'\)'
t_LBRACE = r'\{'
t_RBRACE = r'\}'
t_COMMA = r','
t_SEMI = r';'
t_ASSIGN = r'='

# Caracteres ignorados
t_ignore = ' \t'

# Regra para comentários
def t_COMMENT(t):
    r'//.*'
    pass

# Regra para IDs e palavras reservadas
def t_ID(t):
    r'[a-zA-Z_][a-zA-Z0-9_]*'
    t.type = reserved.get(t.value, 'ID')
    return t

# Regra para FLOAT
def t_FLOAT(t):
    r'[-+]?\d+\.\d+'
    t.value = float(t.value)
    return t

# Regra para INT
def t_INT(t):
    r'[-+]?\d+'
    t.value = int(t.value)
    return t

# Regra para STRING
def t_STRING(t):
    r'\"([^\\\"]|\\.)*\" | \'([^\\\']|\\.)*\''
    return t

# Regra para nova linha
def t_newline(t):
    r'\n+'
    t.lexer.lineno += len(t.value)

# Regra de erro
def t_error(t):
    print(f"Illegal character '{t.value[0]}' at line {t.lexer.lineno}")
    t.lexer.skip(1)

# Construção do lexer
__file__ = "JOJULUC.ipynb"
lexer = lex.lex()

# Ler código do arquivo input.jjl
with open('input.jjl', 'r') as file:
    data = file.read()

# Lexer
lexer.input(data)

# Teste do lexer
#while True:
#    tok = lexer.token()
#    if not tok:
#        break  # No more input
#    print(tok)
```

#Análise Sintática:

```python
import ply.yacc as yacc

# Regras de parsing
precedence = (
    ('left', 'OR'),
    ('left', 'AND'),
    ('left', 'EQ', 'NE'),
    ('left', 'GT', 'LT', 'GE', 'LE'),
    ('left', 'PLUS', 'MINUS'),
    ('left', 'TIMES', 'DIVIDE', 'MOD'),
    ('right', 'NOT'),
)

def p_program(p):
    '''program : functions main_function'''
    p[0] = ('program', p[1], p[2])

def p_functions(p):
    '''functions : function functions
                 | empty'''
    if len(p) == 3:
        p[0] = [p[1]] + p[2]
    else:
        p[0] = []

def p_function(p):
    '''function : FUNCTION TYPE ID LPAREN parameters RPAREN block'''
    p[0] = ('function', p[2], p[3], p[5], p[7])

def p_parameters(p):
    '''parameters : parameter COMMA parameters
                  | parameter
                  | empty'''
    if len(p) == 2:
        p[0] = [p[1]]
    else:
        p[0] = [p[1]] + p[3]

def p_parameter(p):
    '''parameter : TYPE ID'''
    p[0] = ('parameter', p[1], p[2])

def p_main_function(p):
    '''main_function : MAIN LPAREN parameters RPAREN block'''
    p[0] = ('main_function', p[3], p[5])

def p_block(p):
    '''block : LBRACE statements RBRACE'''
    p[0] = ('block', p[2])

def p_statements(p):
    '''statements : statement statements
                  | empty'''
    if len(p) == 3:
        p[0] = [p[1]] + p[2]
    else:
        p[0] = []

def p_statement(p):
    '''statement : declaration_statement
                 | assignment_statement
                 | if_statement
                 | loop_statement
                 | break_statement
                 | return_statement
                 | print_statement'''
    p[0] = ('statement', p[1])

def p_declaration_statement(p):
    '''declaration_statement : TYPE ID ASSIGN expression SEMI'''
    p[0] = ('declare', p[1], p[2], p[4])

def p_assignment_statement(p):
    '''assignment_statement : ID ASSIGN expression SEMI'''
    p[0] = ('assign', p[1], p[3])

def p_if_statement(p):
    '''if_statement : IF LPAREN expression RPAREN block else_statement'''
    p[0] = ('if', p[3], p[5], p[6])

def p_else_statement(p):
    '''else_statement : ELSE block
                      | ELSE if_statement
                      | empty'''
    if len(p) == 3:
        p[0] = ('else', p[2])
    elif len(p) == 4:
        p[0] = ('else', p[3])
    else:
        p[0] = None

def p_loop_statement(p):
    '''loop_statement : LOOP block'''
    p[0] = ('loop', p[2])

def p_break_statement(p):
    '''break_statement : BREAK SEMI'''
    p[0] = ('break',)

def p_return_statement(p):
    '''return_statement : RETURN expression SEMI'''
    p[0] = ('return', p[2])

def p_print_statement(p):
    '''print_statement : PRINT LPAREN expression RPAREN SEMI'''
    p[0] = ('print', p[3])

def p_expression(p):
    '''expression : expression PLUS expression
                  | expression MINUS expression
                  | expression TIMES expression
                  | expression DIVIDE expression
                  | expression MOD expression
                  | expression GT expression
                  | expression LT expression
                  | expression EQ expression
                  | expression NE expression
                  | expression LE expression
                  | expression GE expression
                  | expression AND expression
                  | expression OR expression
                  | NOT expression
                  | LPAREN expression RPAREN
                  | ID LPAREN arguments RPAREN
                  | ID
                  | INT
                  | FLOAT
                  | STRING
                  | BOOL'''
    if len(p) == 4:
        if p[2] == '+':
            p[0] = ('+', p[1], p[3])
        elif p[2] == '-':
            p[0] = ('-', p[1], p[3])
        elif p[2] == '*':
            p[0] = ('*', p[1], p[3])
        elif p[2] == '/':
            p[0] = ('/', p[1], p[3])
        elif p[2] == '%':
            p[0] = ('%', p[1], p[3])
        elif p[2] == '>':
            p[0] = ('>', p[1], p[3])
        elif p[2] == '<':
            p[0] = ('<', p[1], p[3])
        elif p[2] == '==':
            p[0] = ('==', p[1], p[3])
        elif p[2] == '!=':
            p[0] = ('!=', p[1], p[3])
        elif p[2] == '<=':
            p[0] = ('<=', p[1], p[3])
        elif p[2] == '>=':
            p[0] = ('>=', p[1], p[3])
        elif p[2] == '&&':
            p[0] = ('&&', p[1], p[3])
        elif p[2] == '||':
            p[0] = ('||', p[1], p[3])
        else:
            p[0] = p[2]
    elif len(p) == 3:
        p[0] = ('!', p[2])
    elif len(p) == 5:
        p[0] = ('call', p[1], p[3])
    elif len(p) == 2:
        if isinstance(p[1], str):
            if (p[1][0] == '"' and p[1][-1] == '"') or (p[1][0] == "'" and p[1][-1] == "'"):
                p[0] = ('string', p[1])
            elif p[1] == 'true':
                p[0] = ('bool', True)
            elif p[1] == 'false':
                p[0] = ('bool', False)
            else:
                p[0] = ('id', p[1])
        elif isinstance(p[1], int):
            p[0] = ('int', p[1])
        elif isinstance(p[1], float):
            p[0] = ('float', p[1])
        else:
            p[0] = None

def p_arguments(p):
    '''arguments : argument COMMA arguments
                 | argument
                 | empty'''
    if len(p) == 2:
        p[0] = [p[1]]
    else:
        p[0] = [p[1]] + p[3]

def p_argument(p):
    '''argument : ID LPAREN arguments RPAREN
                | ID
                | INT
                | FLOAT
                | STRING
                | BOOL'''
    if len(p) == 5:
        p[0] = ('argument', ('call', p[1], p[3]))
    elif len(p) == 2:
        if isinstance(p[1], str):
            if (p[1][0] == '"' and p[1][-1] == '"') or (p[1][0] == "'" and p[1][-1] == "'"):
                p[0] = ('string', p[1])
            elif p[1] == 'true':
                p[0] = ('bool', True)
            elif p[1] == 'false':
                p[0] = ('bool', False)
            else:
                p[0] = ('id', p[1])
        elif isinstance(p[1], int):
            p[0] = ('int', p[1])
        elif isinstance(p[1], float):
            p[0] = ('float', p[1])

def p_empty(p):
    'empty :'
    pass

def p_error(p):
    if p == None:
        token = "end of file"
    else:
        token = f"{p.type}({p.value}) on line {p.lineno}"
    print(f"Syntax error: Unexpected {token}")

# Construção do parser
parser = yacc.yacc(write_tables=False)

# Ler código no arquivo input.jjl
with open('input.jjl', 'r') as file:
    data = file.read()

# Parse
ast = parser.parse(data)
print(ast)
```

#Análise Semântica e Geração de Código:

```python
import llvmlite.ir as ir
import llvmlite.binding as llvm

# Classe de geração da LLVM IR
class CodeGenerator:
    def __init__(self):
        self.module = ir.Module(name='jojulu')
        self.builder = None
        self.func_symtab = {}
        self.block = None
        self.block_stack = []

    def generate_code(self, ast):
        self.visit(ast)

    def visit(self, node):
        if isinstance(node, tuple):
            if node[0] in ['+', '-', '*', '/', '%', '>', '<', '==', '!=', '<=', '>=', '&&', '||', '!']:
                method_name = 'visit_expression'
            else:
                method_name = f'visit_{node[0]}'
            method = getattr(self, method_name, self.generic_visit)
            return method(node)
        return node

    def generic_visit(self, node):
        raise Exception(f'No visit_{node[0]} method')

    def visit_program(self, node):
        _, functions, main_function = node
        print("PROGRAM")
        for func in functions:
            self.visit(func)
        self.visit(main_function)

    def visit_functions(self, node):
        if node:
            for func in node:
                self.visit(func)

    def visit_function(self, node):
        _, ret_type, name, params, body = node
        print("FUNCTION")
        ret_type = self.get_llvm_type(ret_type)
        param_types = [self.get_llvm_type(param[1]) for param in params if param is not None] if params else []
        func_type = ir.FunctionType(ret_type, param_types)
        function = ir.Function(self.module, func_type, name=name)
        block = function.append_basic_block(name="entry")
        self.builder = ir.IRBuilder(block)
        self.func_symtab[name] = function
        print(f"UPDATED WITH '{name}'")
        if params:
            for i, param in enumerate(params):
                if param is not None:
                    param_name = param[2]
                    param_llvm = function.args[i]
                    self.func_symtab[param_name] = param_llvm
                    print(f"UPDATED WITH '{param_name}'")
        self.visit(body)
        if ret_type == ir.VoidType():
            self.builder.ret_void()

    def visit_main_function(self, node):
        _, params, body = node
        print("MAIN FUNCTION")
        ret_type = ir.VoidType()
        param_types = [self.get_llvm_type(param[1]) for param in params if param is not None] if params else []
        func_type = ir.FunctionType(ret_type, param_types)
        function = ir.Function(self.module, func_type, name="main")
        block = function.append_basic_block(name="entry")
        self.builder = ir.IRBuilder(block)
        self.func_symtab['main'] = function
        if params:
            for i, param in enumerate(params):
                if param is not None:
                    param_name = param[2]
                    param_llvm = function.args[i]
                    self.func_symtab[param_name] = param_llvm
                    print(f"UPDATED WITH '{param_name}'")
        self.visit(body)
        self.builder.ret_void()

    def visit_block(self, node):
        _, statements = node
        for stmt in statements:
            self.visit(stmt[1])

    def visit_declare(self, node):
        _, var_type, name, value = node
        llvm_type = self.get_llvm_type(var_type)
        alloca = self.builder.alloca(llvm_type, name=name)
        value_result = self.visit(value)
        value_type = value_result.type
        if llvm_type == ir.DoubleType() and isinstance(value_type, ir.IntType):
            value_result = self.builder.sitofp(value_result, ir.DoubleType())
        self.builder.store(value_result, alloca)
        self.func_symtab[name] = alloca
        print(f"UPDATED WITH '{name}'")

    def visit_assign(self, node):
        _, name, value = node
        alloca = self.func_symtab[name]
        value_result = self.visit(value)
        value_type = value_result.type
        if alloca.type.pointee == ir.DoubleType() and isinstance(value_type, ir.IntType):
            value_result = self.builder.sitofp(value_result, ir.DoubleType())
        self.builder.store(value_result, alloca)

    def visit_return(self, node):
        _, value = node
        print("RETURN")
        self.builder.ret(self.visit(value))

    def visit_print(self, node):
        _, value = node
        val = self.visit(value)
        printf = self.module.globals.get('printf')
        if not printf:
            voidptr_ty = ir.IntType(8).as_pointer()
            printf_ty = ir.FunctionType(ir.IntType(32), [voidptr_ty], var_arg=True)
            printf = ir.Function(self.module, printf_ty, name='printf')
        if isinstance(val.type, ir.DoubleType):
            fmt = "%f\n\0"
            fmt_name = 'fmt_float'
        elif isinstance(val.type, ir.IntType):
            fmt = "%d\n\0"
            fmt_name = 'fmt_int'
        elif isinstance(val.type, ir.PointerType) and val.type.pointee == ir.IntType(8):
            fmt = "%s\n\0"
            fmt_name = 'fmt_str'
        else:
            raise TypeError(f"Unsupported type for print: '{val.type}'")
        fmt_global = self.module.globals.get(fmt_name)
        if not fmt_global:
            fmt_global = ir.GlobalVariable(self.module, ir.ArrayType(ir.IntType(8), len(fmt)), name=fmt_name)
            fmt_global.initializer = ir.Constant(ir.ArrayType(ir.IntType(8), len(fmt)), bytearray(fmt.encode('utf-8')))
            fmt_global.global_constant = True
        fmt_ptr = self.builder.bitcast(fmt_global, ir.IntType(8).as_pointer())
        if isinstance(val.type, ir.IntType) and val.type.width == 1:
            val = self.builder.zext(val, ir.IntType(32))
        self.builder.call(printf, [fmt_ptr, val])
        print(f"PRINT: '{val}'")

    def visit_expression(self, node):
        if len(node) == 4 and node[1] == '(':
            return self.visit(node[2])
        elif len(node) == 3:
            left = self.visit(node[1])
            right = self.visit(node[2])
            op = node[0]
            if isinstance(left.type, ir.DoubleType) or isinstance(right.type, ir.DoubleType):
                if isinstance(left.type, ir.IntType):
                    left = self.builder.sitofp(left, ir.DoubleType())
                if isinstance(right.type, ir.IntType):
                    right = self.builder.sitofp(right, ir.DoubleType())
            if op == '+':
                if isinstance(left.type, ir.IntType) and isinstance(right.type, ir.IntType):
                    return self.builder.add(left, right)
                else:
                    return self.builder.fadd(left, right)
            elif op == '-':
                if isinstance(left.type, ir.IntType) and isinstance(right.type, ir.IntType):
                    return self.builder.sub(left, right)
                else:
                    return self.builder.fsub(left, right)
            elif op == '*':
                if isinstance(left.type, ir.IntType) and isinstance(right.type, ir.IntType):
                    return self.builder.mul(left, right)
                else:
                    return self.builder.fmul(left, right)
            elif op == '/':
                if isinstance(left.type, ir.IntType) and isinstance(right.type, ir.IntType):
                    return self.builder.sdiv(left, right)
                else:
                    return self.builder.fdiv(left, right)
            elif op == '%':
                if isinstance(left.type, ir.IntType) and isinstance(right.type, ir.IntType):
                    return self.builder.srem(left, right)
                else: return self.builder.frem(left, right)
            elif op == '>':
                return self.builder.icmp_signed('>', left, right)
            elif op == '<':
                return self.builder.icmp_signed('<', left, right)
            elif op == '==':
                return self.builder.icmp_signed('==', left, right)
            elif op == '!=':
                return self.builder.icmp_signed('!=', left, right)
            elif op == '<=':
                return self.builder.icmp_signed('<=', left, right)
            elif op == '>=':
                return self.builder.icmp_signed('>=', left, right)
            elif op == '&&':
                return self.builder.and_(left, right)
            elif op == '||':
                return self.builder.or_(left, right)
        elif len(node) == 2:
            if node[0] == '-':
                return self.builder.neg(self.visit(node[1]))
            elif node[0] == '!':
                return self.builder.not_(self.visit(node[1]))
        else:
            return self.visit(node[1])

    def visit_int(self, node):
        return ir.Constant(ir.IntType(32), node[1])

    def visit_float(self, node):
        return ir.Constant(ir.DoubleType(), node[1])

    def visit_id(self, node):
        var = self.func_symtab[node[1]]
        if isinstance(var.type, ir.PointerType):
            return self.builder.load(var)
        else:
            return var

    def visit_bool(self, node):
        return ir.Constant(ir.IntType(1), node[1])

    def visit_string(self, node):
        _, value = node
        value_without_quotes = value[1:-1]
        global_name = f"str_const_{len(self.module.globals)}"
        string_const = ir.Constant(ir.ArrayType(ir.IntType(8), len(value_without_quotes) + 1), bytearray(value_without_quotes.encode("utf8")) + b'\0')
        global_str = ir.GlobalVariable(self.module, string_const.type, name=global_name)
        global_str.initializer = string_const
        global_str.global_constant = True
        global_str.linkage = 'internal'
        str_ptr = self.builder.bitcast(global_str, ir.IntType(8).as_pointer())
        return str_ptr

    def get_llvm_type(self, jojulu_type):
        #print(jojulu_type)
        if jojulu_type == 'int':
            return ir.IntType(32)
        elif jojulu_type == 'float':
            return ir.DoubleType()
        elif jojulu_type == 'void':
            return ir.VoidType()
        elif jojulu_type == 'bool':
            return ir.IntType(1)
        elif jojulu_type == 'string':
            return ir.PointerType(ir.IntType(8))
        else:
            raise TypeError(f'Unknown type: {jojulu_type}')

    def visit_parameters(self, node):
        if node:
            for param in node:
                self.visit(param)

    def visit_parameter(self, node):
        _, param_type, param_name = node
        llvm_type = self.get_llvm_type(param_type)
        return llvm_type, param_name

    def visit_arguments(self, node):
        if not node:
            return []
        return [self.visit(arg) for arg in node]

    def visit_argument(self, node):
        if node[1] == 'id':
            return self.func_symtab[node[2]]
        elif node[1] == 'int':
            return ir.Constant(ir.IntType(32), node[2])
        elif node[1] == 'float':
            return ir.Constant(ir.DoubleType(), node[2])
        elif node[1] == 'string':
            return ir.Constant(ir.PointerType(ir.IntType(8)), node[2])
        elif node[1] == 'bool':
            return ir.Constant(ir.IntType(1), node[2])
        elif node[1] == 'call':
            return self.func_symtab[node[2]]
        else:
            raise TypeError(f'Unknown argument type: {node[1]}')

    def visit_if(self, node):
        _, condition, then_block, else_block = node
        print("IF")
        cond_val = self.visit(condition)
        then_bb = self.builder.append_basic_block("if_then")
        else_bb = self.builder.append_basic_block("if_else")
        after_if_bb = self.builder.append_basic_block("after_if")
        self.builder.cbranch(cond_val, then_bb, else_bb)
        self.builder.position_at_end(then_bb)
        self.visit(then_block)
        if not self.builder.block.is_terminated:
            self.builder.branch(after_if_bb)
        self.builder.position_at_end(else_bb)
        if else_block:
            self.visit(else_block)
        if not self.builder.block.is_terminated:
            self.builder.branch(after_if_bb)
        self.builder.position_at_end(after_if_bb)

    def visit_else(self, node):
        print("ELSE")
        if node:
            self.visit(node[1])

    def visit_call(self, node):
        _, func_name, args = node
        print("FUNCTION CALL")
        func = self.func_symtab.get(func_name)
        if func is None:
            raise Exception(f"Function '{func_name}' not defined.")
        arg_vals = [self.visit(arg) for arg in args]
        return self.builder.call(func, arg_vals)

    def visit_loop(self, node):
        _, body = node
        print("LOOP")
        loop_start_block = self.builder.append_basic_block("loop_start")
        loop_body_block = self.builder.append_basic_block("loop_body")
        after_loop_block = self.builder.append_basic_block("after_loop")
        self.block_stack.append(after_loop_block)
        self.builder.branch(loop_start_block)
        self.builder.position_at_end(loop_start_block)
        cond = self.visit(('bool', True))
        self.builder.cbranch(cond, loop_body_block, after_loop_block)
        self.builder.position_at_end(loop_body_block)
        self.visit(body)
        if not self.builder.block.is_terminated:
            self.builder.branch(loop_start_block)
        self.builder.position_at_end(after_loop_block)
        self.block_stack.pop()

    def visit_break(self, node):
        print("BREAK")
        if not self.block_stack:
            raise Exception("Break statement outside of loop")
        self.builder.branch(self.block_stack[-1])

# Geração da LLVM IR:
code_generator = CodeGenerator()
code_generator.generate_code(ast)

llvm_ir = str(code_generator.module)

print()
print(llvm_ir)

# Gravação da LLVM IR para o arquivo output.ll
with open("output.ll", "w") as f:
    f.write(llvm_ir)
```
